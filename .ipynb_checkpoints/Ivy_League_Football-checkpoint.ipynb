```python
%matplotlib notebook
from IPython.display import Image
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
```

# 1. Scrape Data from Wikipedia

I will be scraping Ivy League football game data from Wikipedia using BeautifulSoup.

```python
wikipedia_page = 'List_of_Ivy_League_football_standings'
wikipedia_url = 'https://en.wikipedia.org/wiki/{}'.format(wikipedia_page)
```

The tables look like the following, and there is a table for every year from 1956 to 2019.

```python
Image(filename='sample_table.png', width=350, height=350)
```

```python
# Send http/1.1 request using requests
page = requests.get(wikipedia_url)
# Extract table data using BeautifulSoup and lxml's parser
soup = BeautifulSoup(page.content, 'lxml')
tables = soup.find_all('table', {'class': 'infobox'})
```

```python
# Create dataframe recording number of wins of each college
df = pd.DataFrame(columns = ['Brown', 'Columbia', 'Cornell', 'Dartmouth', 
                             'Harvard', 'Penn', 'Princeton', 'Yale'])
year = 1956
# Go through every table from 1956 to 2019
for table in tables:
    school_names = []
    num_of_wins = []
    
    # Go through each row in the table
    for i in range(8):
        # The html <tr> element defines a row in a table, 
        # and the <td> element defines a data cell.
        # Here, I skip the first three rows, which are headings 
        # and don't contain information we are interested in.
        
        ### Go through the first coloum to record college names
        for child in table.find_all('tr')[3:][i].find_all('td')[0].children:
            # Sometimes college names are preceeded by AP ranking, 
            # for example, with a format of 'No. 14 Yale'
            
            ### Skip 'No.' and number
            if child.string.strip() == 'No.':
                continue
            if child.string.strip().isdigit():
                continue
            school_names.append(child.string.strip())
            break
        
        ### Go through the W column of conf to record number of wins
        for child in table.find_all('tr')[3:][i].find_all('td')[1].children:
            if not child.string.isspace():
                num_of_wins.append(child.string[0])
            # Due to the slight different table formatting after 1996
            else:
                for child in table.find_all('tr')[3:][i].find_all('td')[2].children:
                    num_of_wins.append(child.string[0])
    
    # Record in dataframe
    df_y = pd.DataFrame(dict(zip(school_names, num_of_wins)), index=[year])
    df = pd.concat([df, df_y])
    year += 1

df = df.astype(int)            
```

Now I have a dataframe `df` recording the number of wins for each college every year.

# 2. Interactive Visualization

The college performances fluctuate significantly from one year to another. Thus, it would be clearer to use moving average for visualization. Here, I choose to use 10-year moving average of the win percentage.

```python
mov_avg = df.rolling(10, axis=0).mean()/7
mov_avg.iloc[11:16]
```

```python
# Plot data
fig, ax = plt.subplots(figsize=(8,5))
fig.subplots_adjust(right=0.8)
# Picker=True enables pick events
# Pickradius defines the tolerance in points for pick events
mov_avg.plot(legend=False, picker=True, pickradius=5, linewidth=2, ax=ax)
plt.xlim([1965,2019])
plt.ylim([0,1])
# Format y-axis label
ax.yaxis.set_major_formatter(mtick.PercentFormatter(1.0))
ax.set_xlabel('Season')
ax.set_ylabel('10-Year Moving Average Win')
plt.title('Ivy League Football Teams Win %');

### Define pick event
def on_pick(event):
    # Get school name for label box
    school_name = str(event.artist).split('(')[1][:-1]
    # Get school color for highlight
    c = event.artist.get_color()
    
    # Plot data again
    plt.cla()
    # Add an alpha of 0.2 to the whole plot
    mov_avg.plot(legend=False, picker=True, pickradius=5, linewidth=2, ax=ax, alpha=0.2)
    plt.xlim([1965,2019])
    plt.ylim([0,1])
    ax.yaxis.set_major_formatter(mtick.PercentFormatter(1.0))
    ax.set_xlabel('Season')
    ax.set_ylabel('10-Year Moving Average Win')
    plt.title('Ivy League Football Teams Win %')
    # Plot the highlighted line again with no alpha
    line = mov_avg[school_name].plot(legend=False, picker=True, pickradius=5, 
                                     linewidth=2, ax=ax, color=c)
    
    # Label school name next to where the line ends at 2019
    y = list(mov_avg[school_name])[-1]
    plt.text(1.02, y, school_name, va='center', ha="left", 
             bbox=dict(facecolor="w",alpha=0.5),
             transform=ax.get_yaxis_transform(), fontsize=14)
    fig.subplots_adjust(right=0.8)
    
### Connect the pick event to the figure canvas
g = plt.gcf().canvas.mpl_connect('pick_event', on_pick)
```
